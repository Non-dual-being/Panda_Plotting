```python


import pandas as pd
import itertools as it

path4 =r"C:\Users\Stichting GeoFort\Documents\0. Coding\G. GeoFort\totaal.xlsx"

df = pd.read_excel(path4)


        
```

```python
GroupedCat = df.groupby(by='categorie')


```

```python
#zorg voor unieke kolomnamen bij de merge
dataframes = []
for c, dataframe in GroupedCat:
    dataframe.columns = [f"{col}_{c}" for col in dataframe.columns]
    dataframes.append(dataframe)


dataframes[0]
```

```python
merged = pd.merge(dataframes[0], dataframes[1], how='cross')

merged2 = pd.merge(merged, dataframes[2], how ='cross')

merged3 = pd.merge(merged2, dataframes[3], how ='cross')

merged4 = pd.merge(merged3, dataframes[4], how ='cross')

merged5 = pd.merge(merged4, dataframes[5], how ='cross')

cols=pd.Series(merged5.columns)

print(cols)
```

```python
for col in merged5.columns:
    if "co2" in col:  
        merged5[col] = merged5[col].astype(int) 
```

```python
merged5.head()
```

```python
co2_kolommen = [col for col in merged5.columns if "co2" in col]

co2_2500 = []

for index, row in merged5.iterrows():
    co2_som = sum(row[col] for col in co2_kolommen) 

    if co2_som == 2500:
        co2_2500.append(row)

```

```python
co2_2500[1]
```

```python
# Converteren van de lijst van Series objecten naar een nieuwe DataFrame
co2_2500 = pd.DataFrame(co2_2500).reset_index(drop=True)

co2_2500.head()
```
