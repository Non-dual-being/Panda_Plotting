```python
import matplotlib.pyplot  as plt

kleuren = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
# Definieer markertypes voor elk kenmerk paar
markers = [('o', 'Sepal'), ('s', 'Petal')]

# Maak een figuur en as object aan
plt.figure(figsize=(10, 8))

for (marker, kenmerk_type) in markers:
    for species, group in iris.groupby('species'):
        if kenmerk_type == 'Sepal':
            plt.scatter(group['sepal_length'], group['sepal_width'], 
                        color=kleuren[species], alpha=0.5, label=f'{species} {kenmerk_type}', marker=marker)
        else:
            plt.scatter(group['petal_length'], group['petal_width'], 
                        color=kleuren[species], alpha=0.5, label=f'{species} {kenmerk_type}', marker=marker)

# Verbeter de legenda
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))  # Verwijder duplicaten
plt.legend(by_label.values(), by_label.keys(), title="Species & Kenmerk")

plt.title('Iris Dataset Kenmerken per Soort')
plt.xlabel('Length')
plt.ylabel('Width')
plt.grid(True)
plt.show()
```
